```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
```

# Singles Spectra

```python
#Arguments: starting index, ending index, and step size
#Read_function loads a range of CSV files as output by HistToCSV.C as NxN Numpy arrays
#Returns array of data images and array of labels 

def singles_read_function(start_limit, end_limit, step):
    
    #Generate Empty Lists
    list_d=[]
    list_l=[]
    
    #Read Data Files (1D CSV from Root Hist) in given range, reformats to NxN bins 
    i=start_limit
    while i<=end_limit:
        data=open('path/to/data/2dplot_{0}_keVgamma.csv'.format(i), 'r')        
        print('got data', i)
        data=data.read().split(',')
        data=data[:len(data)-1]
        length_data=int(np.sqrt(len(data)))
        data=np.reshape(data, [length_data, length_data])
        data2=np.zeros([length_data, length_data])
        for j in range(1, length_data):
            data2[:, length_data-j]=data[:, j-1]
        data2=data2.astype(np.float16)
        
        list_d.append(data2)
        i+=step
        
    #Load Labels (Already NxN)
    k=start_limit
    while k<=end_limit:
        label=pd.read_csv('path/to/labels/SingleGammaLabels_{0}_keVgamma.csv'.format(k))
        print('got label', k)
        del label['Unnamed: 0']
        label=label.to_numpy()
        label=label.astype(np.float16)
        list_l.append(label)
        k+=step
    
    #Create 1D array with all NxN data 'images'
    array_d=np.asarray(list_d)
    array_l=np.asarray(list_l)
    
    print('formatting')
    
    #Expand 'images' into NxNx1
    array_d=np.expand_dims(array_d, axis=3)
    array_l=np.expand_dims(array_l, axis=3)
    
    
    
        
    
    
    
    return array_d, array_l
        
```

```python
singles_data, singles_labels=singles_read_function(50,9999,1)
```

```python
print(singles_data.shape, singles_labels.shape)
```

```python
newdata=[]
newlabels=[]


for i in range(len(singles_data)):
    
    #Add 12 rows/columns to top, right edges of each image
    newdata.append(np.pad(singles_data[i], ((12,0),(0,12),(0,0)), 'minimum'))
    newlabels.append(np.pad(singles_labels[i], ((12,0),(0,12),(0,0)), 'minimum'))
    

singles_data=np.array(newdata)
singles_labels=np.array(newlabels)

print(singles_data.shape, singles_labels.shape)
            
            
```

# Combining, splitting to train/test

```python
print(singles_data.shape, doubles_data.shape)
```

```python
data=np.concatenate((singles_data, doubles_data), axis=0)
labels=np.concatenate((singles_labels, doubles_labels), axis=0)

print(data.shape, labels.shape)
```

```python
data_train, data_test, labels_train, labels_test=train_test_split(data, labels, test_size=.2)

print(data_train.shape, data_test.shape, labels_train.shape, labels_test.shape)
```

# Standardizing

```python
data_train=np.float64(data_train)

mean=np.mean(data_train)
std=np.std(data_train)

print(mean, std)

data_train=(data_train-mean)/std

print(np.mean(data_train), np.std(data_train))

data_train=np.float16(data_train)
```
